# Gaussian vs Bilateral Filter

This program is used to blur an input image with a Gaussian filter and a bilateral filter.
This is used in the Beleg to demonstrate the operating principles of both methods.

```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(os.path.join('Images', 'test.jpg'), cv2.IMREAD_COLOR)
```

```python
%matplotlib inline
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_gaussian = cv2.GaussianBlur(img, (11,11), 100, 100)
%matplotlib inline
plt.imshow(cv2.cvtColor(img_gaussian, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_bilateral = cv2.bilateralFilter(img, 11, 75,75)
%matplotlib inline
plt.imshow(cv2.cvtColor(img_bilateral, cv2.COLOR_BGR2RGB))
plt.show()
```
