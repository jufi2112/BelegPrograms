```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
img1 = cv2.imread('adam_mmae_vs_mae.png', cv2.IMREAD_COLOR)
img2 = cv2.imread('sgd_mmae_vs_mae.png', cv2.IMREAD_COLOR)
img3 = cv2.imread('rms_mmae_vs_mae.png', cv2.IMREAD_COLOR)
```

```python
print(str(img1.shape) + ' ' + str(img1.dtype))
print(str(img2.shape) + ' ' + str(img2.dtype))
print(str(img3.shape) + ' ' + str(img3.dtype))
```

```python
buffer = np.ones(shape=(323,32,3)).astype(np.uint8)*255
img1 = np.concatenate((img1, buffer), axis=1)
buffer = np.ones(shape=(323,47,3)).astype(np.uint8)*255
img2 = np.concatenate((img2, buffer), axis=1)
print(img1.shape)
print(img2.shape)
print(img3.shape)
```

```python
comb = np.concatenate((img1,img2,img3), axis=0)
```

```python
%matplotlib inline
plt.imshow(comb)
plt.show()
```

```python
cv2.imwrite('combination_mmae_vs_mae.png', comb)
```
