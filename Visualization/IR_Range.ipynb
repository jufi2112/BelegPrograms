```python
import cv2
import numpy as np
import os
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
path = "/media/julien/Transcend/Beleg/data/validation/Infrared"
```

```python
avail_files = os.listdir(path)
files = [file for file in avail_files if Path(file).suffix == '.png']
```

```python
files
```

```python
maximum = []
minimum = []
for file in files:
    img = cv2.imread(os.path.join(path, file), cv2.IMREAD_ANYDEPTH)
    maximum.append(np.amax(img))
    minimum.append(np.amin(img))
```

```python
scale='linear'
title1='Histogram Infrared'
title2='Title2'

file = files[241]
img = cv2.imread(os.path.join(path, file), cv2.IMREAD_ANYDEPTH)

plt.style.use('default')
#fig, (ax1, ax2) = plt.subplots(2)
fig, ax1 = plt.subplots()
x = []
values = 0
if img.dtype == np.uint16:
    x = np.arange(65536)
    values=65536
elif img.dtype == np.uint8:
    x = np.arange(256)
    values=256
else:
    print("Unsupported image depth!")
y = cv2.calcHist(img, [0], None, [values], [0,values])
#fig.suptitle(title)
ax1.plot(x, y.ravel())
ax1.set(ylabel='Occurences', yscale=scale)#, xlim=[50000,60000], ylim=[0,4])
ax1.set_title(title1)
ax1.label_outer()
#ax2.plot(x, y.ravel())
#ax2.set(xlabel='Values', ylabel='Occurences', yscale=scale)
#ax2.set_title(title2)
plt.savefig('histogram_ir')
```
